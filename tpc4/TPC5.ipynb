```python
#Cinema = [Sala]
#Sala = (nlugares, Vendidos, filme)
#nlugares = Int
#filme = String 
#Vendidos = [Int]
#Sala1, Sala2, Sala3. Sala4, Sala5, Sala6, Sala7, Sala8, Sala9, Sala10

Sala1 = (300, [], "SPIDER MAN") 
Sala2 = (150, [], "Mínions 2" )
Sala3 = (280, [], "Impossible")
Sala4 = (200, [], "Mais que vencedores")
Sala5 = (230, [], "Twilight")
Sala6 = (150, [], "Mundo Jurássico")
Sala7 = (200, [], "Miracles from Heaven")

CINEMA = [Sala1,Sala2]

#mostra todos os filmes em exibição
def listar(CINEMA):
    print("\tFILMES EM EXIBIÇÃO:")
    for filme in CINEMA:
        nlugares,ocupados,filmex=filme
        print("\t-",filmex)
    return


#mostra se o lugar escolhido para aquele filme está disponivel
def disponivel(CINEMA,f,l):
    
    for i in CINEMA:   
        available = False, True                             
        nlugares, ocupados, filme = i
        if f.upper() == filme.upper():
            if int(l) <= nlugares:
                if l not in ocupados:
                    available = True
                    print(available,"\ Lugar reservado!")
                else: 
                    available = False
                    print(available,"\ Esse lugar encontra-se ocupado.")
                return available
            else:
                print("Esse lugar não existe nesta sala!")
        else:
            print("O filme inserido não existe neste Cinema! Confira a seguir os filmes em Cartaz.")
            print(listar(CINEMA))
            
    return CINEMA

def vendebilhete(CINEMA, f, l):
    if disponivel(CINEMA, f, l):
        for i in CINEMA:
            nlugares,ocupados,filme= i
            if f.upper()==filme.upper() and l not in ocupados and int(l) <= nlugares:
                ocupados.append(l)
                print("lugares ocupados de momento:",ocupados)
                print("Compra efetuada com sucesso!")
    return CINEMA 
        
#mostrar quantos lugares estão disponíveis ainda para cada filme
def listardisponibilidade(CINEMA):
    for sala in CINEMA:
        lotacao, ocupados, filme= sala 
        num = lotacao - len(ocupados)
        print(filme,"-",num)
    return

#inserirnovasala
def inserirSala(CINEMA,SALA): 
    if SALA in CINEMA:
        print("Esta sala já existe neste cinema!")
        print("O cinema está assim:", CINEMA)
    else:      #em caso de pequenos erros na introdução do nome do filme(erros de maiusculas/minusculas e espaços)
        if SALA.upper().replace(" ", "") == "SALA3":
            CINEMA.append(Sala3)
            print("A sala foi introduzida")
            print("O cinema agora está assim:", CINEMA)
        elif SALA.upper().replace(" ", "") == "SALA4":
            CINEMA.append(Sala4)
            print("A sala foi introduzida")
            print("O cinema agora está assim:", CINEMA)
        elif SALA.upper().replace(" ", "") == "SALA5":
            CINEMA.append(Sala5)
            print("A sala foi introduzida")
            print("O cinema agora está assim:", CINEMA)
        elif SALA.upper().replace(" ", "") == "SALA6":
            CINEMA.append(Sala6)
            print("A sala foi introduzida")
            print("O cinema agora está assim:", CINEMA)
        elif SALA.upper().replace(" ", "") == "SALA7":
            CINEMA.append(Sala7)
            print("A sala foi introduzida")
            print("O cinema agora está assim:", CINEMA)
        else: 
            print("Introduza corretamente os dados.") 

    return CINEMA

#MENU:
print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\t1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
tecla=(input("Olá! Selecione a opção que deseja executar do seguinte MENU!"))

while tecla != "0":    
    if tecla == "1": 
        listar(CINEMA)
        print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\t1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
        tecla=(input("Prima uma nova tecla para executar uma nova ação!"))
    if tecla == "2":
        f=(input("digite o nome do filme que deseja ver:"))
        l=(input("digite o lugar que deseja reservar:"))       
        vendebilhete(CINEMA,f,l)
        print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\t1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
        tecla=(input("Prima uma nova tecla para executar uma nova ação!"))
    if tecla == "3":
        listardisponibilidade(CINEMA)
        print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\tn1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
        tecla=(input("Prima uma nova tecla para executar uma nova ação!"))
    if tecla == "4":
        SALA =(input("Qual é a sala que quer inserir?")) 
        inserirSala(CINEMA,SALA)
        print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\t1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
        tecla=(input("Prima uma nova tecla para executar uma nova ação!"))
    if tecla == "5":
        print(CINEMA)
        print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\t1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
        tecla=(input("Prima uma nova tecla para executar uma nova ação!"))
    else:
        print("-------------------------------------------------------------\n\t\t _MENU_\t\t\n\t0) SAIR\n\t1) VER FILMES EM EXIBIÇÃO\n\t2) COMPRAR LUGAR\n\t3) VER TOTAL DE LUGARES DISPONÍVEIS POR FILME\n\t4) ACRESCENTAR SALA NOVA\n\t5) VER COMO ESTÁ TODO O CINEMA\n-------------------------------------------------------------")
        tecla=(input("Valor inválido! Introduza apenas um dos algarismos do MENU!"))
else:
        print("Estamos a guardar as informações! Fim do programa!")
```
